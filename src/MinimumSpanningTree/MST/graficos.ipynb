```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
memoria = {"AVL": pd.DataFrame(), "ArvoreBinaria": pd.DataFrame(), "Heap": pd.DataFrame()}
tempo = {"AVL": pd.DataFrame(), "ArvoreBinaria": pd.DataFrame(), "Heap": pd.DataFrame()}
for nomeArquivo in os.listdir('./results/'):
    if "tempo" in nomeArquivo:
        nomeAlgoritmo = nomeArquivo[0:nomeArquivo.find("_")]
        tempo[nomeAlgoritmo] = pd.concat([tempo[nomeAlgoritmo], pd.read_csv(f"./results/{nomeArquivo}")])
    if "memoria" in nomeArquivo:
        nomeAlgoritmo = nomeArquivo[0:nomeArquivo.find("_")]
        memoria[nomeAlgoritmo] = pd.concat([memoria[nomeAlgoritmo], pd.read_csv(f"./results/{nomeArquivo}")])
for nome, df in memoria.items():
    df.drop('Unnamed: 0', axis=1, inplace=True)
    df.reset_index(inplace=True)
for nome, df in tempo.items():
    df.drop('Unnamed: 0', axis=1, inplace=True)
    df.rename(columns={"Memory": "Time"}, inplace=True)
    df.reset_index(inplace=True)
```

```python
sns.set_style("darkgrid") 
plt.figure(figsize=(100, 60))
plt.tick_params(labelsize=70)
cores = ['black', 'blue', 'green', 'purple']
for i, (algoritmo, df) in enumerate(tempo.items()):
    
    plot = sns.lineplot(x="Nodes", y="Time", data=df, linewidth=10, color=cores[i])
    
labels = {"AVL": 19, "ArvoreBinaria": 7, "Heap": 7}
for i, (key, value) in enumerate(labels.items()):
    plt.annotate(key, (tempo[key]["Nodes"][value], tempo[key]["Time"][value]), fontsize=70,
                 color='black', fontweight='bold', rotation=15)
plt.xscale("log")
plt.ylim(0, 200)
plot.set_title("Tempo nos algoritmos para a AGM", fontsize=120)  # Título
plot.set_xlabel("Quantidade de nós", fontsize=80)  # Rótulo do eixo x
plot.set_ylabel("Tempo", fontsize=80)  # Rótulo do eixo y

plt.show()
```
